Our goal: Do highly rated wines have similar characteristics or features?

- fixed acidity: most acids involved with wine (do not evaporate readily)
- volatile acidity: the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste
- citric acid: found in small quantities, citric acid can add 'freshness' and flavor to wines
- residual sugar: the amount of sugar remaining after fermentation stops
- chlorides: the amount of salt in the wine
- free sulfur dioxide: the free form of SO2, it prevents microbial growth and the oxidation of wine
- total sulfur dioxide: amount of free and bound forms of S02; evident in the nose and taste of wine
- density: the density of water is close to that of water depending on the percent alcohol and sugar content
- pH: describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic)
- sulphates: a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant
- alcohol: the percent alcohol content of the wine
- quality: output variable (based on sensory data, score between 0 and 10)

```python
# check how big our datas are

display(white_df.head(3))
print("white.shape:", white_df.shape, '\n')

display(red_df.head(3))
print("red.shape:", red_df.shape)
```

```python
# find duplicate values and dropping it

print("White wine duplicate:", white_df.duplicated().sum())
white_df = white_df.drop_duplicates().reset_index(drop=True)

print("Red wine duplicate:", red_df.duplicated().sum())
red_df = red_df.drop_duplicates().reset_index(drop=True)
```

```python
# how big after dropping it?
print("white.shape:", white_df.shape)

print("red.shape:", red_df.shape)
```

```python
# We are going to add wine color for visiblity when graphing, then concat

white_df['color'] = 'white'
red_df['color'] = 'red'

wine = pd.concat([white_df, red_df], axis=0)
```

```python
wine.sample(5)
```

```python
# check the distribution of the target variable "quality"

sns.countplot(x='quality', data = wine)
plt.title("Distribution of Wine Quality")
plt.show()
```

we decide, if its equal or more than 7, we consider it as a high(good) quality wine.

If the 'quality' value is greater than or equal to 3 and less than or equal to 6 (3 ≤ quality ≤ 6), it will be labeled as 'Poor'.

If the 'quality' value is greater than 6 and less than or equal to 9 (6 < quality ≤ 9), it will be labeled as 'Good'.

```python
# binning class label in to two categories

bins = [3,6,9]
groups = ['Poor','Good']
wine['Quality?'] = pd.cut(wine['quality'], bins, labels = groups)

sns.countplot(x = 'Quality?', data = wine)
```

```python
# correlation between "quality" and other variables

correlation_matrix = wine.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Brief recap of correlation:
    
Negative Correlation
1) Alcohol with density (-0.78) 
2) citric acid with volatile acidity(-0.38)
3) Alcohol with residual sugar (-0.31)
4) Free sulfur dioxide with volatile acidity(-0.35)
5) Density with Quality(-0.33)

Positive Correlation
1) density with residual sugar(0.52)
2) Alcohol with Quality (0.47)
3) Free sulfur with total sulfur(0.72)
4) Total sulfur dioxide with Residual sugar (0.49)
5) Density with Fixed acidity(0.48)


#### But We want to focus on wine quality!

Most of the features showed little correlation with the wine quality ratings. However, alcohol content appeared to have the highest positive correlation at around 0.47. 
Following that, density was the next significant feature. 
On the other hand, volatile acidity showed a negative correlation with the ratings, indicating that higher values of volatile acidity might lead to a stronger and less desirable sour taste. 

We further investigate how the top four features (alcohol, density, volatile acidity, chlorides)

with relatively higher correlations (correlation coefficient > 0.1) manifest themselves with respect to the wine quality ratings.

### 2. Density

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data=wine, x='density', bins=30, ax=axes[0])
axes[1].plot(wine.index, wine['density'], '.', )
axes[1].set_xlabel('index')
axes[1].set_ylabel('density')
plt.suptitle('Density Distribution')
plt.show()
```

```python
#finding very weird outliers
wine[wine['density']>1.01]
```

```python
# density distribution after removing the outliers

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
filtered = wine.query("density < 1.01")
sns.histplot(data=filtered, x='density', bins=30, ax=axes[0])
sns.boxplot(data=filtered, x='Quality?', y='density', ax=axes[1])
axes[0].set_title('Density Distribution (outlier removed)')
axes[0].tick_params(axis='x', labelrotation=30)
axes[1].set_title('Density vs Quality (outlier removed)')
plt.tight_layout()
plt.show()
```

The density distribution, excluding outliers, follows a thick normal distribution. As observed in the correlation coefficients, the density tends to decrease with increasing wine quality ratings.

However, for wines with a quality rating of 5, the median density slightly increases compared to other ratings. This finding aligns with the negative correlation (-0.67) between alcohol and density.

As alcohol has a lower density than water, higher alcohol content leads to lower overall density. The graph visually confirms that as wine quality ratings increase, alcohol content increases, while density decreases.

## 3. Volatile Acidity


```python
palette = ['#dbf47c','#7e002f']

fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(data=wine, x='volatile acidity', hue='color',
             palette=palette, bins=50)
plt.title('Volatile Acidity Distribution')
plt.show()
```

The distribution of volatile acidity shows a distinct difference depending on the wine color. In white wine, most of the values are concentrated around 0.3, whereas in red wine, the values are relatively evenly distributed from 0.3 to above 0.7.

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5), sharey=True)

for i, color in enumerate(['red', 'white']):
    sns.boxplot(data=wine.query("color==@color"),
                x='Quality?', y='volatile acidity',
                boxprops=dict(alpha=0.7), ax=axes[i])
sns.boxplot(data=wine,
            x='Quality?', y='volatile acidity',
            ax=axes[0])

axes[0].set_title('Red wine')
axes[1].set_title('White wine')
plt.suptitle("Volatile Acidity vs Quality by Color")
plt.show()
```

Due to the distinct difference in the distribution of volatile acidity depending on the wine color, I created separate box plots based on the wine color. For red wine, as the quality rating increases, the amount of volatile acidity decreases. However, for white wine, the difference in volatile acidity among quality ratings is not as pronounced.
correlation coeff of 0.26 doesn't really matter because white wine dataset is three times larger than the red wine dataset and volatile acidity in white wine does not have a strong association with the quality rating, making its impact less significant.

## 4. Chlorides

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(data=wine, x='chlorides', bins=30, ax=axes[0, 0])
sns.histplot(data=wine, x='chlorides', hue='color',
             palette=palette, bins=30, ax=axes[0, 1])

wine_colors = ['white', 'red']
for i, (color, edgecolor) in enumerate(zip(wine_colors, ['white', 'y'])):
    sns.scatterplot(data=wine.query("color==@color"),
                    x=wine.query("color==@color").index, y='chlorides',
                    color=palette[i], ax=axes[1, 0])
    
sns.boxplot(data=wine, x='color', y='chlorides',
            palette=palette, ax=axes[1, 1])

titles = ['Overall', 'By Color', 'Actual values', 'Quantile']
for i, title in enumerate(titles):
    axes[i//2, i%2].set_title(title)
plt.suptitle('Chlorides Distribution')
plt.tight_layout()
plt.show()
```

The majority of values for chloride content fall between 0 and 0.15, with white wine exhibiting lower chloride levels compared to red wine. 

However, outliers are affecting the proper representation of the data. After removing outliers above the 98th percentile (around 110 data points), the graph shows the following distribution.

```python
filtered = wine[wine['chlorides'] < wine['chlorides'].quantile(0.98)]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(data=filtered, x='chlorides', bins=30, ax=axes[0, 0])
sns.histplot(data=filtered, x='chlorides', hue='color',
             palette=palette, bins=30, ax=axes[0, 1])

for i, (color, edgecolor) in enumerate(zip(wine_colors, ['white', 'y'])):
    sns.scatterplot(data=filtered.query("color==@color"),
                    x=filtered.query("color==@color").index, y='chlorides',
                    color=palette[i], ax=axes[1, 0])
    
sns.boxplot(data=filtered, x='color', y='chlorides',
            palette=palette, ax=axes[1, 1])

titles = ['Overall', 'By Color', 'Actual values', 'Quantile']
for i, title in enumerate(titles):
    axes[i//2, i%2].set_title(title)
plt.suptitle('Chlorides Distribution w/o outliers')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for i, df in enumerate([wine, filtered]):
    for j, color in enumerate(wine_colors):
        sns.boxplot(data=df.query("color==@color"), x='quality', y='chlorides',
                    color=palette[j], boxprops=dict(alpha=0.7), ax=axes[i, j])

for i, w_o in enumerate(['w/', 'w/o']):
    for j, color in enumerate(wine_colors):
        axes[i, j].set_title(f"{color.title()} wine {w_o} outliers")
plt.suptitle("Chlorides vs Quality")
plt.tight_layout()
plt.show()
```

## Quality by Color of Wine?

white wine vs red.. which whine has better quality overall?

```python
wine_quality = wine.groupby('color')['Quality?'].value_counts(normalize=True)
wine_quality.name = 'proportion'
wine_quality = wine_quality.reset_index()
```

```python
palette = ['#7e002f','#dbf47c']

fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(data=wine_quality, x='Quality?', y='proportion',
            hue='color', palette=palette,
            edgecolor='k', alpha=0.7, ax=ax)
plt.title('Proportion of Quality by Color')
plt.legend(loc='upper right')
plt.show()
```

```python
# let's view this not only our point of view(7 or greater as good), but also general quality view

wine_quality = wine.groupby('color')['quality'].value_counts(normalize=True)
wine_quality.name = 'proportion'
wine_quality = wine_quality.reset_index()

fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(data=wine_quality, x='quality', y='proportion',
            hue='color', palette=palette,
            edgecolor='k', alpha=0.7, ax=ax)
plt.title('Proportion of Quality by Color')
plt.legend(loc='upper right')
plt.show()
```

```python
wine_quality_diff = wine_quality.pivot(index='quality', columns='color', values='proportion')
wine_quality_diff['prop_diff'] = wine_quality_diff['red'] - wine_quality_diff['white']
custom_palette = wine_quality_diff['prop_diff'].map(lambda x: palette[0] if x > 0 else palette[1])
wine_quality_diff = wine_quality_diff.reset_index()



plt.figure(figsize=(8, 5))
sns.barplot(data=wine_quality_diff, x='quality', y='prop_diff',
            palette=custom_palette, alpha=0.7, edgecolor='k')
plt.legend(['red % - white %'], loc='upper right')
plt.title('Proportion difference between Red/White wine')
```

The majority of wines are classified into quality levels 5, 6, and 7. Among these, red wines are more prevalent in quality level 5 compared to white wines, while white wines are more distributed in quality levels 6 and 7. In the other quality levels 3 and 4, there is a relatively low absolute number of included wines, but the ratio between red and white wines is nearly similar. For the higher quality level 8, white wines are more abundant, similar to quality levels 6 and 7. 

### Conclusion:

As the quality level increases:

- The alcohol content tends to be higher.
- The density decreases.
- In red wines, the level of volatile acidity decreases. (Excessive volatile acidity can lead to an unpleasant sour taste.)
- The chloride ion (element of salt) content is lower, with minimal variations. This change is even less pronounced in white wines.
- There is a slightly higher probability of the wine being a white wine.
